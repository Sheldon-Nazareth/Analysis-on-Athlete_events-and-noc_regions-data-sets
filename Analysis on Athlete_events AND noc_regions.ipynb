```python
###############################################################################################################################
# Answer 1 
refined_data['Medal'].unique()
```

```python
Medal_counts = refined_data['Medal'].value_counts()
Medal_counts
```

```python
#using a groupby function on team we can count the number of Gold medals and sort them for ease. 
Gold_medal_data = refined_data[refined_data['Medal'] == 'Gold'].groupby('Team')['Medal'].count().sort_values(ascending=False)
Gold_medal_data
```

```python
import matplotlib.pyplot as plt
Gold_medal_data_plot = Gold_medal_data.head(3)
Gold_medal_data_plot
```

```python
fig = plt.bar(x=Gold_medal_data_plot.index, height=Gold_medal_data_plot.values)
# adding supplementary formatting 
plt.title('Gold Medalist Countries')
plt.xlabel('Country')
plt.ylabel('Number of Gold Medals Won')
plt.show()
plt.rcParams['figure.figsize'] = [15, 4]
```

```python
# We can see the top 3 countries with the maximum number of Gold medals won during the olympics USA leads by a landslide
```

```python
# For **clarity** these are the total number of male and female candidates segregated on the basis of their gender.
Gender_counts = refined_data['Sex'].value_counts()
Gender_counts
```

```python
# Here the same group data is segregated on the basis of candidates recievng a medal.
Gender_counts = refined_data[refined_data['Medal'] != 'No'].groupby('Sex')['Medal'].count().sort_values(ascending=False)
Gender_counts
```

```python
fig = plt.pie(Gender_counts, labels=Gender_counts.index, autopct='%.2f')
```
